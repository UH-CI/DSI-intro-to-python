````
# i
abo = ['A', 'B', 'O']
abo2 = ['A', 'B', 'O']
bloodtypes = []
for blood in abo:
  for blood2 in abo2:
    bloodtype = blood + blood2
    if sorted(bloodtype) not in bloodtypes:
      bloodtypes.append(sorted(bloodtype))
print(bloodtypes)
  

        
````

````
# ii
abo=['A','B','O']
abo2=['A','B','O']
rhesus=['Rh+','Rh-']
rhesus2=['Rh+','Rh-']
bloodtypes=[]
for blood in abo:
  for blood2 in abo2:
    for rh in rhesus:
      for rh2 in rhesus2:
        bloodtype = [blood, blood2, rh, rh2]
        if sorted(bloodtype) not in bloodtypes:
          bloodtypes.append(sorted(bloodtype))
print(bloodtypes)
````

````
# iii
phenotypes=[]
for genotype in bloodtypes:
  if ('A' in genotype) and ('B' in genotype):
    phenotype=['AB']
  elif 'B' in genotype:
    phenotype=['B']
  elif 'A' in genotype:
    phenotype=['A']
  else:
    phenotype=['O']
  if 'Rh+' in genotype:
    phenotype.append('Rh+')
  else:
    phenotype.append('Rh-')
  phenotypes.append(phenotype)
phenotypes
````

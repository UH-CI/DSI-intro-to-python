** Exercise 1 **

The list below contains tuples with a food's total calories and calories from fat. Use an if statement nested in a for loop to print the names of the food with more than 50% calories from fat.

```
foods=[('oreo', 53, 21),('butter', 800, 792)]
for food in foods:
  if food[2]/food[1]>0.5:
    print(food[0])
```

According to the World Bank, a high income country is defined as a country with a GNI per capita of 12,056 US dollars or more, a upper-middle income country has a GNI per capita between 3,896 to 12,055 US dollars, a lower-middle income country as a GNI per capita between 996 to 3,895 US dollars, and a low income country has a GNI per capita of 995 US dollars or less (2017 GNIs). Write a piece of code that loops through the dictionary and sorts the countries and GNI per capitas into four different dictionaries: `high`, `upper_middle`, `lower_middle` and `low` based on the countries' income classification.

```
high={}
upper_middle={}
lower_middle={}
low={}
for country, gni in countries_gni.items():
  if gni>=12056:
    high[country]=gni
  elif gni>=3896:
    upper_middle[country]=gni
  elif gni>=996:
    lower_middle[country]=gni
  else:
    low[country]=gni
print(high)
print(upper_middle)
print(lower_middle)
print(low)
```

#### Exercise 3
Isograms are words that do not have repeating consecutive or non-consecutive letters. Write a piece of code that determines whether a given word is an isogram or not.

Hint:

```
# isograms
word='dinosaur'
temp={}
for letter in word:
  if letter not in temp:
    temp[letter] = 1
  else:
    print(word + ' is not an isogram.')
    break
  if len(temp) == len(word):
    print(word + " is an isogram.")
```

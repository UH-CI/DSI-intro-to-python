$$
T = \frac{T_{0}}{\sqrt{1-\frac{v^2}{c^2}}}
$$

With $T$ being the time interval experienced by the moving object, $T_{0}$ being the time interval experienced by the staitionary reference, $v$ being the velocity of the moving object, and $c$ being the speed of light.

#### Exercise 5
---
Write a piece of code that tests whether two words are anagrams. Anagrams are words that share the exact same letters, but in a different order. If the words are anagrams, print "`word_1` and `word_2` are anagrams". If the words are not anagrams, print "`word_1` and `word_2` are not anagrams"

Hint: You will need to use the `sorted()` function to achieve this.

```
word_1 = 'binary'
word_2 = 'brainy'
if sorted(word_1) == sorted(word_2):
  print(word_1 + " and " + word_2 + " are anagrams")
else:
  print(word_1 + " and " + word_2 + " are not anagrams")
 
```
